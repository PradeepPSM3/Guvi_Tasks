```python
import re
import pickle

#Registering User with Email and Password

def registerUser():
    Email,Password = getEmail(),getPassWord()
    saveUser(Email,Password)
    pageStart()

#Reading email with required condition 

def getEmail():
     isInValidEmail = True
     while isInValidEmail:
          email = input("Enter the User Email : ")
          emailExp = r'\b[A-Za-z0-9._%+-]+@[A-Za-z.-]+\.[A-Z|a-z]{2,}\b'
          if(re.fullmatch(emailExp,email)):
                isInValidEmail = False
                return email
          else:
            print("Invalid Email try again....")
            
            
#Reading password with required condition 

def getPassWord():
    isInvalidPassword = True
    while isInvalidPassword :
        print("\nPassword must be Alphanumeric  and atleast one special charachter with 6 to 15 in length")
        numeric=upper=lower=special=0
        password = input("Enter a password : ")
        if(15 >= len(password) >= 6):            
            for i in password:
                if i==" ":
                    break
                elif i.isupper():
                    upper+=1
                elif i.isdigit():
                    numeric+=1
                elif i.islower():
                    lower+=1
                else:
                    special+=1
            if (numeric and upper and lower and special):
                isInvalidPassword=False
                return password
            else:
                print("\nPassWord doesn't met the Expectation")
                
                
#User Login with credentials

def loginUser():
    Email,Password = getEmail(),getPassWord()
    createifNoFile()
    with open("UserCredentioal.txt","rb") as fs:
        try:
            credential=fileObject=pickle.load(fs)
            if credential.get(Email) == None:
                print("Email is not Register Yet")
            elif credential.get(Email) == Password:
                print("you are successfully logged")
            else:
                print("invalid Email or Password")
                
        except Exception as ex:
            print(ex)
            
#Forget Password Scenario

def forGetPassword():
    Email=getEmail()
    createifNoFile()
    with open("UserCredentioal.txt","rb") as fs:
        try:
            credential=fileObject=pickle.load(fs)
            if credential.get(Email) == None:
                print("Email is not Register Yet")
                pageStart()
            else:
                print(f'Password for the {Email} is {credential.get(Email)}')
        except:
            print("Invalid User")
            


# Page Startup Press Required key to process information
def pageStart():    
    userInput =input("\nPress 1 --> To Register\nPress 2 --> To Login\nPress 3 --> Forget Password\nPress 0 --> Exit\n Input : ")  
    userInput = userInput.strip()
    if(userInput == "1"):
        registerUser()
    elif userInput == "2":
        loginUser()
    elif userInput == "3":
        forGetPassword()
    elif userInput == "0":
        print("Thanks for Visiting us , Have a Great Day..")

    else :
        print("Invalid Input, Please Try again ..")
        pageStartStart()

#Saving User Credentials in Binary format in  the file

def saveUser(email,password):
    obj={}
    isEmailPresent = False
    isEmptyFile = createifNoFile()
    if not isEmptyFile:        
        with open("UserCredentioal.txt","rb") as fs:
            try:
                obj=pickle.load(fs)
                if obj.get(email) != None:
                    isEmailPresent=True

            except Exception as ex:
                print(ex)
    if not isEmailPresent:        
        with open("UserCredentioal.txt","wb+") as fs:
            obj[email] = password
            pickle.dump(obj,fs)
            print("User Creaated Successfully...")
    else:
        print("Email Alredy Excist , Try different one...")   
        pageStart()

def createifNoFile():
    from os.path import exists
    if not exists("UserCredentioal.txt"):
        fs= open("UserCredentioal.txt","x") 
        fs.close()
        return True
    return False
        
#Starting of the  operation
pageStart()
    
```
